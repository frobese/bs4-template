```python
import re

from bs4 import BeautifulSoup as bs
```

Nach dem Import der ben√∂tigten Bibliotheken k√∂nnen wir das HTML-Dokument √∂ffnen und uns eine formatierte Version des body-tags ausgeben lassen.

```python
with open("./source/frorum.html") as page:
    soup = bs(page, "lxml")

print(soup.body.prettify())
```

Wir durchsuchen jetzt den body des Dokuments nach tags, deren class-Attribut das Wort "content" enth√§lt - Und finden praktischerweise genau eins! 

```python
for tag in soup.body.find_all(class_=re.compile("content")):
    print(tag.attrs)
```

Wir wissen schon, dass alle Eintr√§ge des Flexbox-Layouts ein "card" in der CSS-Klasse stehen haben. Also verfeinern wir die Suche entsprechend.

```python
content = soup.body.find(id="content")
elements = content.find_all(class_=re.compile("card"))
for tag in elements:
    print(tag.attrs)
```

Die ersten beiden Elemente sehen anders aus als der Rest. Schauen wir uns diese Elemente n√§her an:

```python
for tag in elements[:1]:
    print(tag.prettify())
```

Die beiden ersten Eintr√§ge waren nur ein Wrapper und die √úberschrift - f√ºr uns also nicht interessant. Vom Rest der Liste k√∂nnen wir uns jetzt die √úberschriften ausgeben lassen. Das sind alle Blogeintr√§ge, perfekt!

```python
posts = elements[2:]
for tag in posts:
    print(tag.h5.string)
```

Zus√§tzlich zu den Titeln der Artikel wollen wir auch die URLs zum jeweiligen Volltext finden und abspeichern. Dazu erweitern wir das Skript:

```python
posts = elements[2:]
for tag in posts:
    title = tag.h5
    link = title.parent.get("href")
    print(title.string, link)
```
